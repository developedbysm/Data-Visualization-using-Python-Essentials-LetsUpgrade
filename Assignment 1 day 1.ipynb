## Question 1

### Importing libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.arange(40,50)
b=np.arange(50,60)
```

### Creating Line Plot

```python
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot")
plt.plot(a,b,'b--.',markersize=20,linewidth=2,markerfacecolor='green')
plt.grid()
plt.show()
```

## Question 2

```python
days = [1,2,3,4,5,6,7] 
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70, 90, 160, 150, 140, 145, 175]
```

### Creating Line Plot For sales_1 and sales_2 For All Days 

```python
plt.plot(days, sales_1,label='sales_1',marker='o')
plt.plot(days, sales_2,label='sales_2',marker='*')
plt.xlabel('Days of The Week',color='grey')
plt.ylabel('sales',color='grey')
plt.title("Days vs Sales",color='grey')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```

```python
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]
```

## Creating Subplot

```python
fig,axs=plt.subplots(2,3,figsize=(20,5))
axs[0,0].plot(x,y1)
axs[0,1].plot(x,y2)
axs[0,2].plot(x,y3)
axs[1,0].plot(x,y4)
axs[1,1].plot(x,y5)
```
